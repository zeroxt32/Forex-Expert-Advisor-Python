```python
import talib
import numpy as np

# Example data
high_prices = np.array([10, 12, 11, 14, 13])
low_prices = np.array([9, 10, 10, 11, 12])
close_prices = np.array([11, 11, 10, 13, 13])

# Calculate True Range (TR)
tr = np.maximum(high_prices - low_prices, np.abs(high_prices - np.roll(close_prices, 1)), np.abs(low_prices - np.roll(close_prices, 1)))

# Calculate Standard Deviation of True Range (SDTR)
sdtr = talib.STDDEV(tr, timeperiod=14)

# Calculate Average True Range (ATR)
atr = talib.ATR(high_prices, low_prices, close_prices, timeperiod=14)

# Calculate Relative Volatility Index (RVI)
rvi = 100 * (atr / sdtr)

print(rvi)
```

```python
import numpy as np
import os
from tensorflow import keras
from ncps import wirings
from ncps.tf import LTC
import pandas as pd
import matplotlib.pyplot as plt
```

```python
scaled_df = pd.read_csv("scaled_df.csv")
scaled_df.head(40)
```

```python
y = np.array( [0] * 60 + [1] * 50)
y = y.reshape(1,len(y), 1)
y.shape
```

```python
stacked_feature = np.expand_dims(stacked_feature, axis = 0)
stacked_feature.shape
```

```python
wiring = wirings.AutoNCP(8,1) # 8 neurons in total, 1 output (motor neuron)

model = keras.models.Sequential(
    [
        keras.layers.InputLayer(input_shape=(None, 160)),
        # here we could potentially add layers before and after the LTC network
        LTC(wiring, return_sequences=True),
    ]
)
model.compile(
    optimizer=keras.optimizers.Adam(0.01), loss='mean_squared_error'
)

model.summary()
hist = model.fit(x=stacked_feature, y=y, batch_size=1, epochs=300,verbose=1)
```
